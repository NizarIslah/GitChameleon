```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

# Assume 'model_dfs' is a list of dataframes split by model
# Reconstruct the full DataFrame with all models' data
full_df = pd.concat(model_dfs, ignore_index=True)

# Verify the reconstruction
print(f"Reconstructed DataFrame shape: {full_df.shape}")

# Create a sample_id for aggregation
# Assuming each sample is represented in each of the 10 models, sample_id should be within range [0, 96]
full_df['sample_id'] = full_df.index // n_models

# Multiply values by 100
full_df['pass_at_10'] *= 100
full_df['feedback_pass_at_10'] *= 100

# Aggregate the mean pass_at_10 and feedback_pass_at_10 values by sample_id
agg_df = full_df.groupby('sample_id').agg({
    'pass_at_10': 'mean',
    'feedback_pass_at_10': 'mean'
}).reset_index()

# Number of unique samples
n_samples = len(agg_df)
print(f"Number of unique samples: {n_samples}")

# Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(18, 9))

# Plot histogram for pass_at_10
hist_pass_at_10, bins_pass_at_10, patches_pass_at_10 = axes[0].hist(agg_df['pass_at_10'].dropna(), bins=10, range=(0, 100), edgecolor='black', color="thistle", linewidth=3)
axes[0].set_title('Difficulty of samples (Baseline)', fontsize=25)
axes[0].set_xlabel('pass@10', fontsize=25)
axes[0].set_ylabel('Frequency', fontsize=25)
axes[0].tick_params(axis='both', which='major', labelsize=20)
axes[0].yaxis.set_major_locator(MaxNLocator(integer=True))

# Calculate the maximum y-axis limit based on the first subplot
max_y_limit = max(hist_pass_at_10) * 1.1

# Plot histogram for feedback_pass_at_10
hist_feedback_pass_at_10, bins_feedback_pass_at_10, patches_feedback_pass_at_10 = axes[1].hist(agg_df['feedback_pass_at_10'].dropna(), bins=10, range=(0, 100), edgecolor='white', color="purple", linewidth=3)
axes[1].set_title('Difficulty of samples (+ Error Feedback)', fontsize=25)
axes[1].set_xlabel('pass@10', fontsize=25)
#axes[1].set_ylabel('Number of samples', fontsize=25)
axes[1].tick_params(axis='both', which='major', labelsize=20)

# Apply the same y-axis limit and ticks to both subplots
axes[0].set_ylim(0, max_y_limit)
axes[1].set_ylim(0, max_y_limit)

# Set y-axis ticks to be the same for both subplots
# Extract y-ticks from the first subplot
y_ticks = axes[0].get_yticks()
axes[1].set_yticks(y_ticks[:-1])

# Adjust layout
plt.tight_layout()

# Save plot
plt.savefig("sample_diff_hist.pdf", bbox_inches='tight')

# Show plot
plt.show()

# Sanity check: print the total frequency of each histogram
print(f"Total number of samples: {n_samples}")

# Calculate and print the total frequency from histograms
print(f"Total frequency for pass_at_10 histogram: {sum(hist_pass_at_10)}")
print(f"Total frequency for feedback_pass_at_10 histogram: {sum(hist_feedback_pass_at_10)}")

# Ensure the total frequencies match the number of samples
assert abs(sum(hist_pass_at_10) - n_samples) < 1e-6, "Mismatch in pass_at_10 histogram frequency"
assert abs(sum(hist_feedback_pass_at_10) - n_samples) < 1e-6, "Mismatch in feedback_pass_at_10 histogram frequency"
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'all_samples_data' is already loaded into df
df = all_samples_data.copy()

# Convert columns to numeric to handle any potential issues with types
for col in ['pass_at_10', 'pass_at_10.1', 'feedback_pass_at_10', 'feedback_pass_at_10.1']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# List of models to include
models_of_interest = [
    'CodeLlama-34b-instruct-hf',
    'Codestral-22B-v0.1-hf',
    'gpt4o'
]

# Create a 3x2 subplot figure
fig, axes = plt.subplots(3, 2, figsize=(20, 20), sharex='col', sharey='row')

# Dictionary to store frequencies for sanity check
frequencies = {}

# Loop through each model and plot histograms
for i, model in enumerate(models_of_interest):
    # Filter the dataframe for the current model
    model_df = df[df['Unnamed: 0'] == model]
    if model == 'CodeLlama-34b-instruct-hf':
        m = 'CodeLLama-34B'
    elif model == 'Codestral-22B-v0.1-hf':
        m = 'Codestral-22B'
    else:
        m = 'GPT-4o'

    model_df['pass_at_10'] *= 100
    model_df['feedback_pass_at_10'] *= 100
    
    # Plot baseline pass_at_10 histogram
    hist_pass_at_10, bins_pass_at_10, patches_pass_at_10 = axes[i, 0].hist(model_df['pass_at_10'].dropna(), bins=3, range=(0, 100), edgecolor='black', color="thistle", linewidth=3)
    axes[i, 0].set_title(f'{m} (Baseline)', fontsize=40)
    axes[i, 0].set_xlabel('pass@10 (%)', fontsize=35)
    axes[i, 0].set_ylabel('Frequency', fontsize=35)
    axes[i, 0].tick_params(axis='both', which='major', labelsize=35)
    
    # Annotate the bars with frequency counts
    for patch, count in zip(patches_pass_at_10, hist_pass_at_10):
        height = patch.get_height()
        axes[i, 0].text(patch.get_x() + patch.get_width() / 2, height, f'{int(count)}', ha='center', va='bottom', fontsize=30, color='black')
    
    # Adjust y-axis limits to ensure annotations fit
    axes[i, 0].set_ylim(0, max(hist_pass_at_10) * 1.15)
    
    # Plot feedback_pass_at_10 histogram
    hist_feedback_pass_at_10, bins_feedback_pass_at_10, patches_feedback_pass_at_10 = axes[i, 1].hist(model_df['feedback_pass_at_10'].dropna(), bins=3, range=(0, 100), edgecolor='white', color="purple", linewidth=3)
    axes[i, 1].set_title(f'{m} (+ Error Feedback)', fontsize=40)
    axes[i, 1].set_xlabel('pass@10 (%)', fontsize=35)
    #axes[i, 1].set_ylabel('Frequency', fontsize=35)
    axes[i, 1].tick_params(axis='both', which='major', labelsize=35)
    
    # Annotate the bars with frequency counts
    for patch, count in zip(patches_feedback_pass_at_10, hist_feedback_pass_at_10):
        height = patch.get_height()
        axes[i, 1].text(patch.get_x() + patch.get_width() / 2, height, f'{int(count)}', ha='center', va='bottom', fontsize=30, color='black')
    
    # Adjust y-axis limits to ensure annotations fit
    #axes[i, 1].set_ylim(0, max(hist_feedback_pass_at_10) * 1.1)
    
    # Manually set x-ticks
    x_tick_values = [0, 50, 100]
    for ax in axes.flat:
        ax.set_xticks(x_tick_values)
    
    # Store frequencies for sanity check
    frequencies[model] = {
        'pass_at_10': sum(hist_pass_at_10),
        'feedback_pass_at_10': sum(hist_feedback_pass_at_10)
    }

# Adjust layout
plt.tight_layout()
plt.savefig("model_diff_hist.pdf", bbox_inches='tight')

# Show plot
plt.show()

# Sanity check: print the total frequency of each histogram
for model in models_of_interest:
    print(f"Model: {model}")
    print(f"Total frequency for pass_at_10 histogram: {frequencies[model]['pass_at_10']}")
    print(f"Total frequency for feedback_pass_at_10 histogram: {frequencies[model]['feedback_pass_at_10']}")
    print(f"Expected total frequency: 97")
    print()
```

```python
import matplotlib.pyplot as plt

# Release dates as a list of strings
release_dates = [
    "Jun 15, 2021", "Oct 21, 2021", "Oct 28, 2022", "March 15, 2023", "Oct  3, 2021",
    "Feb 28, 2021", "April  9, 2022", "Feb 28, 2022", "August 22, 2020",
    "Oct 16, 2019", "Jan 30, 2016", "September 20, 2017", "June 21, 2014", "Jun 27, 2020",
    "Feb 12, 2015", "Mar 30, 2023", "July  6, 2023", "May 16, 2022", "April 14, 2022",
    "March  3, 2023", "Jan 30, 2023", "Dec 21, 2022", "May 12, 2022", "Dec  8, 2022",
    "Jun 30, 2023", "Sept 24, 2021", "March  4, 2020", "May 18, 2019", "Mar 26, 2021",
    "Sept 16, 2022", "Nov 16, 2021", "May 15, 2024", "Mar 18, 2023"
]

# Extract years from release dates
years = [int(date.split(",")[-1].strip()) for date in release_dates]

# Count the occurrences of each year
year_counts = {}
for year in years:
    if year not in year_counts:
        year_counts[year] = 0
    year_counts[year] += 1

# Define years range (2014 to 2024)
years_range = range(2014, 2024)

# Extract counts for years in the range
year_counts_in_range = [year_counts.get(year, 0) for year in years_range]

# Define the width of each group
bar_width = 0.75

# Plot
fig, axs = plt.subplots(1, 1, figsize=(12, 8), constrained_layout=True)

# First subplot
axs.bar(years_range, year_counts_in_range, width=bar_width, color='royalblue', alpha=0.9, edgecolor="k")

# Select every alternate year for x-axis labels
alternate_years = years_range[::2]  # Slice with step size 2

# Set x-axis labels and rotation for better readability
plt.xticks(alternate_years, fontsize=30) 
axs.yaxis.set_tick_params(labelsize=35)
#axs.set_title('PACS', fontsize=45)
#axs.set_ylim(0,0.5)
# set x axis label
axs.set_xlabel('Version Release Year', fontsize=30)
axs.set_ylabel('Sample Count', fontsize=30)


plt.tight_layout()
plt.grid(color='grey', linestyle='--', linewidth=1, alpha=0.5)

plt.savefig('version_year.pdf', bbox_inches='tight')
plt.show()
```
